## **What is the impact of discount on sales?**

```python
df['Discount'].value_counts()
```

```python
# Sales vs. Discount Ratio
discount_per_sales =df.groupby(['Discount'])['Sales'].sum().reset_index()
discount_per_sales
```

```python
# Create a bar chart illustrating the relationship between sales and discounts
plt.figure(figsize=(15,6))
sns.barplot(data=discount_per_sales, x='Discount', y='Sales', palette='icefire')
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.title('Sales vs Discount')
plt.tick_params(axis='x', rotation=45)
plt.show()
```

```python
# Visualize the relationship between Sales and Discount using a scatter plot.
plt.scatter(df['Discount'] ,df['Sales'])
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.title('Sales vs Discount Relationship')

plt.show()
```

- It's clear that higher discount percentages generally lead to lower sales amounts. This aligns with common retail practices, where deeper discounts can attract more customers but might result in lower overall revenue.

- The data appears to represent the total sales amount for different discount levels. The sales amounts range from around `$5,000` to over` $1 million`. The discount ratios also vary from 0% (no discount) to 80%.
